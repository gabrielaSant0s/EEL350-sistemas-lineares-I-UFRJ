````
t = np.arange(0, 4, 0.001)
````

````
def serieF2(numHarmo):
    x = np.zeros(t.shape)

    for k in range(1, (2 * numHarmo) + 1, 2):
        x += (4/ k*np.pi)*(1-(-1)**k) * np.sin(((np.pi**2)/4)*k*t)
    
    return x
````

````
def serieF1(numHarmo):
    x = np.zeros(t.shape)

    for k in range(1, (2 * numHarmo) + 1, 2):
        x += (2/ k*np.pi)*(1-(-1)**k) * np.sin(((np.pi**2)/2)*k*t)
    
    return x
````
